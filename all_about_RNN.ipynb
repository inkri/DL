## Time Series forecasting

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
import matplotlib.pyplot as plt

# Generate dummy stock price data
num_days = 1000
starting_price = 100
random_walk = np.cumsum(np.random.normal(0, 1, num_days))
stock_prices = starting_price + 10 * random_walk
dates = pd.date_range(start='2020-01-01', periods=num_days)
stock_data = pd.DataFrame({'Date': dates, 'Close': stock_prices})

# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(stock_data['Close'].values.reshape(-1, 1))

# Define window size (number of previous days' closing prices to use for prediction)
window_size = 10

# Create input and target data
X, y = [], []
for i in range(len(scaled_data) - window_size):
    X.append(scaled_data[i:i+window_size, 0])
    y.append(scaled_data[i+window_size, 0])
X, y = np.array(X), np.array(y)

# Reshape data for RNN input (samples, time steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test, y_train, y_test = X[:train_size], X[train_size:], y[:train_size], y[train_size:]

# Define the RNN model
model = Sequential()
model.add(SimpleRNN(units=50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=2)

# Plot training loss and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Make predictions
predictions = model.predict(X_test)

# Plot actual vs. predicted prices
plt.plot(scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual')
plt.plot(scaler.inverse_transform(predictions), label='Predicted')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock Price Prediction')
plt.legend()
plt.show()
```

### Embedding:
#### Embedding is a way to represent words as dense vectors in a continuous vector space.
#### Word embeddings are learned during the training of a neural network.
#### Words with similar meanings are expected to have similar vector representations.

### Key Differences:

#### Nature of Representation:
#### Integer encoding represents words with discrete integer values.
#### Embedding represents words as continuous, dense vectors.

#### Learnability:
#### Integer encoding is typically fixed and doesn't adapt during training.
#### Embedding vectors are learned during the training of a neural network.

#### Semantic Relationships:
#### Embedding is designed to capture semantic relationships between words.
#### Integer encoding does not inherently capture semantic relationships.

#### Dimensionality:
#### Integer encoding assigns a unique integer to each word without considering relationships.
#### Embedding provides a lower-dimensional representation that captures semantic relationships.

```python
# Integer Encoding Example

import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense, Reshape
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Dummy data
texts = [
    "This is a positive sentence.",
    "I love working with deep learning models.",
    "Negative reviews are not helpful.",
    "The movie was terrible; I hated it.",
]

labels = [1, 1, 0, 0]

# Tokenize the text for integer encoding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
total_words = len(tokenizer.word_index) + 1

# Integer encode the text
sequences = tokenizer.texts_to_sequences(texts)
x_integer = pad_sequences(sequences)

# Convert labels to NumPy array
y_labels = np.asarray(labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_integer, y_labels, test_size=0.2, random_state=42, shuffle=True)

# Manually split the training set into training and validation sets
validation_split = 0.2
split_index = int((1 - validation_split) * len(x_train))

x_train_split, x_val, y_train_split, y_val = (
    x_train[:split_index],
    x_train[split_index:],
    y_train[:split_index],
    y_train[split_index:],
)

# Build the model with Integer Encoding
integer_model = Sequential()
integer_model.add(Embedding(input_dim=total_words, output_dim=8, input_length=x_train.shape[1]))
integer_model.add(Reshape((x_train.shape[1], 8)))  # Reshape to add the missing dimension
integer_model.add(SimpleRNN(4))
integer_model.add(Dense(1, activation='sigmoid'))
integer_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with Integer Encoding
integer_model.fit(x_train_split, y_train_split, epochs=5, batch_size=2, validation_data=(x_val, y_val), shuffle=True)

# Evaluate the model on the test set
integer_predictions = (integer_model.predict(x_test) > 0.5).astype('int32')
integer_accuracy = accuracy_score(y_test, integer_predictions)

print(f"Accuracy with Integer Encoding: {integer_accuracy * 100:.2f}%")
```

```python
# Embedding Example

import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Dummy data
texts = [
    "This is a positive sentence.",
    "I love working with deep learning models.",
    "Negative reviews are not helpful.",
    "The movie was terrible; I hated it.",
]

labels = [1, 1, 0, 0]

# Tokenize the text for embedding
max_words = 10000
maxlen = 20
embedding_dim = 8

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
x_embedded = pad_sequences(sequences, maxlen=maxlen)

# Convert labels to NumPy array
y_labels = np.asarray(labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_embedded, y_labels, test_size=0.2, random_state=42, shuffle=True)

# Build the model with Embedding
embedding_model = Sequential()
embedding_model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=maxlen))
embedding_model.add(SimpleRNN(4))
embedding_model.add(Dense(1, activation='sigmoid'))
embedding_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with Embedding
embedding_model.fit(x_train, y_train, epochs=5, batch_size=2, validation_split=0.2, shuffle=True)

# Evaluate the model on the test set
embedding_predictions = (embedding_model.predict(x_test) > 0.5).astype('int32')
embedding_accuracy = accuracy_score(y_test, embedding_predictions)

print(f"Accuracy with Embedding: {embedding_accuracy * 100:.2f}%")
```

# Many-to-One RNN:

## In a Many-to-One RNN, the input sequence is of variable length, and the output is a fixed-size prediction. Here's an example using a simple LSTM layer:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

#In this example, the input shape is (None, 10), where None represents variable sequence length, and 10 is the number of features in each time step.
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10)))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# Many-to-Many RNN:

## In a Many-to-Many RNN, both the input and output sequences have variable lengths. Here's an example using the TimeDistributed layer to apply a Dense layer at each time step:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed

#In this example, return_sequences=True in the LSTM layer ensures that it returns the full sequence of outputs for each input sequence.
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10), return_sequences=True))
model.add(TimeDistributed(Dense(1, activation='sigmoid')))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# One-to-Many RNN:

## In a One-to-Many RNN, the input is a fixed-size sequence, and the output is a variable-length sequence. Here's an example using RepeatVector:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, RepeatVector

#In this example, the RepeatVector layer repeats the output of the previous layer to match the desired output sequence length.
output_sequence_length=10
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10)))
model.add(RepeatVector(output_sequence_length))  # Set output_sequence_length accordingly
model.add(LSTM(64, return_sequences=True))
model.add(TimeDistributed(Dense(1, activation='sigmoid')))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# Problems with RNN

Recurrent Neural Networks (RNNs) have been widely used for sequential data modeling, but they also have some limitations and challenges:

Vanishing and Exploding Gradients: RNNs suffer from vanishing and exploding gradient problems during training. This occurs because gradients are multiplied across many time steps during backpropagation, causing them to either become very small (vanishing gradients) or very large (exploding gradients). This can make it difficult for the model to learn long-term dependencies.

Difficulty in Capturing Long-Term Dependencies: RNNs have difficulty in capturing long-term dependencies in sequences. This is because information from earlier time steps may gradually diminish as it propagates through the network, making it hard for the model to remember information from distant past time steps.

Limited Contextual Information: RNNs process sequences one element at a time, which limits their ability to capture contextual information from the entire sequence. This may result in suboptimal performance, especially when dealing with long sequences.

Computationally Intensive: Training RNNs can be computationally intensive, especially for long sequences and deep architectures. This can lead to longer training times and higher resource requirements.

Fixed-Length Inputs and Outputs: Traditional RNNs require fixed-length inputs and produce fixed-length outputs, which can be restrictive for tasks involving variable-length sequences.

Difficulty in Parallelization: Due to their sequential nature, RNNs are difficult to parallelize efficiently across time steps, limiting their scalability on hardware accelerators like GPUs.

Difficulty with Training Stability: Training RNNs can be challenging to stabilize, especially for deep architectures and long sequences. Various techniques such as gradient clipping, batch normalization, and careful weight initialization are often required to address this issue.

Despite these challenges, RNNs have been foundational in sequential data modeling and have paved the way for more advanced architectures like Long Short-Term Memory (LSTM) networks and Gated Recurrent Unit (GRU) networks, which aim to address some of these limitations. Additionally, attention mechanisms and transformer architectures have emerged as powerful alternatives for sequential data processing in recent years.








# Long Short-Term Memory (LSTM)

### Long Short-Term Memory (LSTM) is a type of recurrent neural network (RNN) architecture designed to overcome the vanishing gradient problem that occurs in traditional RNNs. LSTMs are particularly well-suited for tasks involving sequential data, such as time series prediction, speech recognition, natural language processing, and more.

### LSTM Architecture:
#### LSTM is a type of recurrent neural network (RNN) designed to address the vanishing gradient problem encountered in traditional RNNs. The vanishing gradient problem occurs when gradients diminish exponentially as they are propagated back through time during training, making it challenging for RNNs to learn long-range dependencies.

#### The LSTM architecture consists of several key components:

#### Memory Cell (Ct): This is the main component of the LSTM and is responsible for maintaining long-term dependencies. It can selectively remember or forget information over time.

#### Input Gate (it): This gate controls the flow of new information into the memory cell. It decides which values from the input should be updated and stored in the memory cell.

#### Forget Gate (ft): This gate controls the flow of information from the previous cell state (Ct-1) to the current cell state (Ct). It decides which information from the previous state should be forgotten or retained.

#### Output Gate (ot): This gate controls the flow of information from the memory cell to the output. It decides which information from the current cell state should be output at the current time step.

How LSTM Works:
Here's a step-by-step explanation of how an LSTM cell processes input:

Forget Gate Operation:

The forget gate takes the previous cell state (Ct-1) and the current input (Xt) as input.
It passes them through a sigmoid activation function, which outputs values between 0 and 1.
The output of the forget gate (ft) determines which information from the previous cell state should be discarded (0) and which should be retained (1).
Input Gate Operation:

The input gate also takes the previous cell state (Ct-1) and the current input (Xt) as input.
It passes them through a sigmoid activation function, which outputs values between 0 and 1, representing the input gate (it).
It also passes the current input (Xt) through a tanh activation function, which outputs values between -1 and 1, representing the candidate values (~Ct).
The output of the input gate and the candidate values are multiplied element-wise, resulting in the values that will be added to the cell state.
Update the Cell State (Ct):

The forget gate output (ft) is used to scale the previous cell state (Ct-1), determining which information should be retained.
The input gate output (it) and the candidate values (~Ct) are used to update the cell state.
The new cell state (Ct) is obtained by adding the scaled previous cell state to the updated values.
Output Gate Operation:

The output gate takes the current input (Xt) and the updated cell state (Ct) as input.
It passes them through a sigmoid activation function, which outputs values between 0 and 1, representing the output gate (ot).
The current cell state (Ct) is passed through a tanh activation function to squash the values between -1 and 1.
The output of the output gate (ot) and the squashed cell state are multiplied element-wise, resulting in the final output of the LSTM cell.
By selectively updating and passing information through the gates, LSTMs can effectively capture long-range dependencies in sequential data, making them suitable for various tasks such as time series prediction, natural language processing, and speech recognition.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Generate some dummy data for demonstration
def generate_data(n_samples, n_timesteps, n_features):
    X = np.random.rand(n_samples, n_timesteps, n_features)
    y = np.sum(X, axis=2)
    return X, y

# Hyperparameters
n_samples = 1000
n_timesteps = 10
n_features = 5
n_units = 50  # Number of LSTM units
batch_size = 32
epochs = 10

# Generate training data
X_train, y_train = generate_data(n_samples, n_timesteps, n_features)

# Define LSTM model
model = Sequential()
model.add(LSTM(n_units, input_shape=(n_timesteps, n_features)))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs)

# Generate some test data for prediction
X_test, y_test = generate_data(10, n_timesteps, n_features)

# Make predictions
predictions = model.predict(X_test)
print("Predictions:", predictions)
print("Actual:", y_test)
```

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Embedding, Dense, Dropout

# Set hyperparameters
max_features = 5000  # Number of words to consider as features
maxlen = 100  # Max length of review (pad shorter ones, truncate longer ones)
embedding_size = 128
lstm_units = 128
dropout_rate = 0.2
batch_size = 64
epochs = 3

# Load the IMDB dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Preprocess the data
X_train = pad_sequences(X_train, maxlen=maxlen)
X_test = pad_sequences(X_test, maxlen=maxlen)

# Define the LSTM model
model = Sequential()
model.add(Embedding(max_features, embedding_size, input_length=maxlen))
model.add(Dropout(dropout_rate))
model.add(LSTM(lstm_units))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test))

# Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1] * 100))
```

# Gated Recurrent Unit

### GRU Architecture:
#### GRU is another type of recurrent neural network (RNN) architecture similar to LSTM, designed to address the vanishing gradient problem and capture long-range dependencies in sequential data. GRU simplifies the architecture of LSTM by merging the forget and input gates into a single update gate.


### The main components of a GRU unit include:

### Update Gate (zt): This gate controls the flow of information from the previous hidden state to the current hidden state. It decides how much of the previous hidden state to retain and how much of the new candidate values to incorporate.

### Reset Gate (rt): This gate controls the extent to which the previous hidden state should influence the calculation of the new candidate values. It decides which parts of the previous hidden state are relevant for the current time step.

### Candidate Hidden State (~ht): This represents the new candidate values that could be added to the hidden state.

### Hidden State (ht): This is the output of the GRU unit, representing the updated hidden state for the current time step.

### How GRU Works:
#### Here's a step-by-step explanation of how a GRU unit processes input:

### Reset Gate Operation:
#### The reset gate takes the previous hidden state (ht-1) and the current input (xt) as input.
#### It passes them through a sigmoid activation function, which outputs values between 0 and 1, representing the reset gate (rt).
#### The output of the reset gate (rt) determines which parts of the previous hidden state should be reset or retained.

### Update Gate Operation:
#### The update gate takes the previous hidden state (ht-1) and the current input (xt) as input.
#### It passes them through a sigmoid activation function, which outputs values between 0 and 1, representing the update gate (zt).
#### The output of the update gate (zt) determines how much of the previous hidden state should be retained (1) and how much of the new candidate values (~ht) should be incorporated (0).


### Candidate Hidden State Calculation:
#### The candidate hidden state (~ht) is calculated based on the reset gate output (rt) and the current input (xt).
#### It combines the previous hidden state (ht-1) and the current input (xt) using the reset gate output (rt) to determine which parts of the previous hidden state are relevant for the current time step.

### Update Hidden State:
#### The updated hidden state (ht) is calculated by blending the previous hidden state (ht-1) and the candidate hidden state (~ht) based on the update gate output (zt).
#### If the update gate output (zt) is close to 1, more information from the previous hidden state is retained. If it is close to 0, more information from the candidate hidden state is incorporated.


#### By using the reset and update gates, GRU units can effectively capture dependencies in sequential data while simplifying the architecture compared to LSTM. This makes GRUs computationally more efficient and easier to train in some cases. They are widely used in various tasks such as natural language processing, speech recognition, and machine translation.

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Embedding, Dense, Dropout

# Set hyperparameters
max_features = 5000  # Number of words to consider as features
maxlen = 100  # Max length of review (pad shorter ones, truncate longer ones)
embedding_size = 128
gru_units = 128
dropout_rate = 0.2
batch_size = 64
epochs = 3

# Load the IMDB dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Preprocess the data
X_train = pad_sequences(X_train, maxlen=maxlen)
X_test = pad_sequences(X_test, maxlen=maxlen)

# Define the GRU model
model = Sequential()
model.add(Embedding(max_features, embedding_size, input_length=maxlen))
model.add(Dropout(dropout_rate))
model.add(GRU(gru_units))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test))

# Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1] * 100))
```

# Deep RNNs | Stacked RNNs | Stacked LSTMs | Stacked GRUs

### Deep RNNs, Stacked RNNs, Stacked LSTMs, and Stacked GRUs are variations of recurrent neural network (RNN) architectures that aim to capture complex patterns and long-term dependencies in sequential data. Let's briefly explain each of these concepts:

### Deep RNNs:
#### Deep RNNs refer to recurrent neural networks with multiple layers of recurrent units. Each layer processes the input sequence sequentially, and the output of one layer serves as the input to the next layer.
#### Deep RNNs allow for more complex representations of sequential data by learning hierarchical features at different levels of abstraction.

### Stacked RNNs:
#### Stacked RNNs are similar to deep RNNs, but they stack multiple recurrent layers on top of each other without connections between layers. Each layer in a stacked RNN processes the entire sequence independently and passes its output to the layer above.
#### Stacked RNNs can learn more complex temporal patterns compared to single-layer RNNs, as each layer extracts features at different levels of temporal abstraction.

### Stacked LSTMs:
#### Stacked LSTMs are a variation of deep RNNs where the recurrent layers are comprised of LSTM (Long Short-Term Memory) units. Like stacked RNNs, each layer in a stacked LSTM processes the entire sequence independently.
#### Stacked LSTMs are particularly effective in capturing long-term dependencies in sequential data and mitigating the vanishing gradient problem commonly encountered in deep networks.

### Stacked GRUs:
#### Stacked GRUs are similar to stacked LSTMs, but they use GRU (Gated Recurrent Unit) units instead of LSTM units. GRUs have a simpler architecture compared to LSTMs, consisting of fewer parameters, which can lead to faster training times.
#### Stacked GRUs are also effective in capturing long-term dependencies and are suitable for tasks such as sequence modeling and natural language processing.

    
#### In summary, deep RNNs, stacked RNNs, stacked LSTMs, and stacked GRUs are advanced architectures of recurrent neural networks that allow for the modeling of complex temporal relationships in sequential data. Each architecture has its advantages and is suited to different tasks depending on factors such as computational resources, training data, and the nature of the problem being solved. Experimentation and empirical evaluation are crucial for determining the most effective architecture for a given task.

# Bidirectional RNN | BiLSTM | Bidirectional LSTM | Bidirectional GRU

### Bidirectional RNNs, BiLSTM, Bidirectional LSTM, and Bidirectional GRU are variants of recurrent neural network (RNN) architectures designed to capture bidirectional dependencies in sequential data. Let's discuss each of these concepts:

### Bidirectional RNN (BRNN):
#### A Bidirectional RNN consists of two RNNs: one processing the input sequence in the forward direction and the other processing the input sequence in the backward direction.
#### The outputs from both directions are typically concatenated or combined in some way to produce the final output.
#### Bidirectional RNNs are effective in capturing dependencies in both past and future contexts, making them useful for tasks such as sequence labeling and sequence classification.

### BiLSTM (Bidirectional LSTM):
#### BiLSTM extends the idea of Bidirectional RNNs by replacing the standard RNN units with LSTM units.
#### Like Bidirectional RNNs, BiLSTM processes the input sequence in both forward and backward directions simultaneously.
#### Each LSTM unit in the forward direction and backward direction maintains its own hidden state and cell state.
#### BiLSTM is particularly effective in capturing long-range dependencies and is widely used in tasks such as natural language processing, sentiment analysis, and named entity recognition.

### Bidirectional LSTM:
#### Bidirectional LSTM is the same as BiLSTM; the terms are often used interchangeably.
#### Bidirectional LSTM combines the advantages of LSTM units (such as memory cell and gates) with bidirectional processing to capture both past and future information.

### Bidirectional GRU:
#### Bidirectional GRU is similar to Bidirectional LSTM, but it uses GRU units instead of LSTM units.
#### GRU units have a simpler architecture compared to LSTM units, consisting of fewer parameters, which can lead to faster training times.
#### Bidirectional GRU is effective in capturing bidirectional dependencies in sequential data and is suitable for tasks such as sequence modeling and machine translation.

### In summary, Bidirectional RNNs, BiLSTM, Bidirectional LSTM, and Bidirectional GRU are architectures that enable the model to consider information from both past and future contexts. These architectures are powerful tools for capturing bidirectional dependencies in sequential data, making them well-suited for a wide range of tasks in natural language processing, time series analysis, and other sequential data processing tasks.

## Bidirectional LSTM (BiLSTM)

```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Embedding, Dense, Bidirectional, Dropout

# Set hyperparameters
max_features = 5000  # Number of words to consider as features
maxlen = 100  # Max length of review (pad shorter ones, truncate longer ones)
embedding_size = 128
lstm_units = 64
dropout_rate = 0.2
batch_size = 64
epochs = 3

# Load the IMDB dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

# Preprocess the data
X_train = pad_sequences(X_train, maxlen=maxlen)
X_test = pad_sequences(X_test, maxlen=maxlen)

# Define the BiLSTM model
model = Sequential()
model.add(Embedding(max_features, embedding_size, input_length=maxlen))
model.add(Dropout(dropout_rate))
model.add(Bidirectional(LSTM(lstm_units)))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_data=(X_test, y_test))

# Evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1] * 100))
```

## Bidirectional LSTM model

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional
from sklearn.preprocessing import MinMaxScaler

# Generate sample time series data (replace with your own data)
def generate_time_series(start_date, end_date):
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')
    data = np.sin(np.arange(len(date_range)) * 2 * np.pi / 365)  # Example: sine wave data
    return pd.Series(data, index=date_range)

# Convert time series data into sequences
def create_sequences(data, sequence_length):
    sequences = []
    for i in range(len(data) - sequence_length):
        sequences.append(data[i:i+sequence_length])
    return np.array(sequences)

# Prepare data
start_date = '2010-01-01'
end_date = '2019-12-31'
sequence_length = 30  # Length of input sequences
data = generate_time_series(start_date, end_date)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))
X = create_sequences(scaled_data, sequence_length)
y = scaled_data[sequence_length:]

# Split data into train and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Define the Bidirectional LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(units=50, activation='relu'), input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32)

# Evaluate the model
train_loss = model.evaluate(X_train, y_train, verbose=0)
test_loss = model.evaluate(X_test, y_test, verbose=0)
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)

# Make predictions
predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)

# Invert predictions to original scale
predictions_train = scaler.inverse_transform(predictions_train)
predictions_test = scaler.inverse_transform(predictions_test)
y_train_inv = scaler.inverse_transform(y_train)
y_test_inv = scaler.inverse_transform(y_test)

# Evaluate performance (e.g., using RMSE)
from sklearn.metrics import mean_squared_error
rmse_train = np.sqrt(mean_squared_error(y_train_inv, predictions_train))
rmse_test = np.sqrt(mean_squared_error(y_test_inv, predictions_test))
print("Train RMSE:", rmse_train)
print("Test RMSE:", rmse_test)
```
