```python
import numpy as np
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
from keras.optimizers import Adam

# Generate some dummy data for training
X_train = np.random.random((1000, 10, 1))  # 1000 sequences of length 10 with 1 feature
y_train = np.random.randint(2, size=(1000,))  # Binary labels
print("Train: ",X_train[0])

# Define the RNN model
model = Sequential()
model.add(SimpleRNN(units=32, input_shape=(10, 1), activation='relu'))  # 32 recurrent units
model.add(Dense(units=1, activation='sigmoid'))  # Output layer for binary classification
print(model.summary())

# Define optimizer and compile the model
optimizer = Adam(lr=0.001)  # Learning rate set to 0.001
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)  # 10 epochs, batch size of 32, 20% validation split

# Evaluate the model
X_test = np.random.random((200, 10, 1))  # 200 sequences of length 10 with 1 feature for testing
y_test = np.random.randint(2, size=(200,))  # Binary labels for testing
loss, accuracy = model.evaluate(X_test, y_test)

print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```

## Time Series forecasting

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
import matplotlib.pyplot as plt

# Generate dummy stock price data
num_days = 1000
starting_price = 100
random_walk = np.cumsum(np.random.normal(0, 1, num_days))
stock_prices = starting_price + 10 * random_walk
dates = pd.date_range(start='2020-01-01', periods=num_days)
stock_data = pd.DataFrame({'Date': dates, 'Close': stock_prices})

# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(stock_data['Close'].values.reshape(-1, 1))

# Define window size (number of previous days' closing prices to use for prediction)
window_size = 10

# Create input and target data
X, y = [], []
for i in range(len(scaled_data) - window_size):
    X.append(scaled_data[i:i+window_size, 0])
    y.append(scaled_data[i+window_size, 0])
X, y = np.array(X), np.array(y)

# Reshape data for RNN input (samples, time steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test, y_train, y_test = X[:train_size], X[train_size:], y[:train_size], y[train_size:]

# Define the RNN model
model = Sequential()
model.add(SimpleRNN(units=50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=2)

# Plot training loss and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Make predictions
predictions = model.predict(X_test)

# Plot actual vs. predicted prices
plt.plot(scaler.inverse_transform(y_test.reshape(-1, 1)), label='Actual')
plt.plot(scaler.inverse_transform(predictions), label='Predicted')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock Price Prediction')
plt.legend()
plt.show()
```

### Integer Encoding:
#### Integer encoding involves representing each word in a text with a unique integer.
#### Words are assigned unique numerical identifiers based on their frequency or other criteria.
#### It's a simple way to convert categorical data (words) into numerical format.

```python
from keras.preprocessing.text import Tokenizer

texts = ["cat", "dog", "cat", "bird", "dog", "dog"]
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
encoded_texts = tokenizer.texts_to_sequences(texts)

print("Original Texts:", texts)
print("Integer Encoded Texts:", encoded_texts)
```

### Embedding:
#### Embedding is a way to represent words as dense vectors in a continuous vector space.
#### Word embeddings are learned during the training of a neural network.
#### Words with similar meanings are expected to have similar vector representations.

```python
from keras.layers import Embedding
import numpy as np

# Assume 'total_words' is the total number of unique words in your dataset
total_words = 10

embedding_dim = 4
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=embedding_dim, input_length=1))

# Suppose 'cat' corresponds to index 1, 'dog' to index 2, etc.
word_indices = np.array([1, 2, 1, 3, 2, 2])

embedded_vectors = model.predict(word_indices)

print("Word Indices:", word_indices)
print("Embedded Vectors:")
print(embedded_vectors)
```

### Key Differences:

#### Nature of Representation:
#### Integer encoding represents words with discrete integer values.
#### Embedding represents words as continuous, dense vectors.

#### Learnability:
#### Integer encoding is typically fixed and doesn't adapt during training.
#### Embedding vectors are learned during the training of a neural network.

#### Semantic Relationships:
#### Embedding is designed to capture semantic relationships between words.
#### Integer encoding does not inherently capture semantic relationships.

#### Dimensionality:
#### Integer encoding assigns a unique integer to each word without considering relationships.
#### Embedding provides a lower-dimensional representation that captures semantic relationships.

```python
# Integer Encoding Example

import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense, Reshape
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Dummy data
texts = [
    "This is a positive sentence.",
    "I love working with deep learning models.",
    "Negative reviews are not helpful.",
    "The movie was terrible; I hated it.",
]

labels = [1, 1, 0, 0]

# Tokenize the text for integer encoding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
total_words = len(tokenizer.word_index) + 1

# Integer encode the text
sequences = tokenizer.texts_to_sequences(texts)
x_integer = pad_sequences(sequences)

# Convert labels to NumPy array
y_labels = np.asarray(labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_integer, y_labels, test_size=0.2, random_state=42, shuffle=True)

# Manually split the training set into training and validation sets
validation_split = 0.2
split_index = int((1 - validation_split) * len(x_train))

x_train_split, x_val, y_train_split, y_val = (
    x_train[:split_index],
    x_train[split_index:],
    y_train[:split_index],
    y_train[split_index:],
)

# Build the model with Integer Encoding
integer_model = Sequential()
integer_model.add(Embedding(input_dim=total_words, output_dim=8, input_length=x_train.shape[1]))
integer_model.add(Reshape((x_train.shape[1], 8)))  # Reshape to add the missing dimension
integer_model.add(SimpleRNN(4))
integer_model.add(Dense(1, activation='sigmoid'))
integer_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with Integer Encoding
integer_model.fit(x_train_split, y_train_split, epochs=5, batch_size=2, validation_data=(x_val, y_val), shuffle=True)

# Evaluate the model on the test set
integer_predictions = (integer_model.predict(x_test) > 0.5).astype('int32')
integer_accuracy = accuracy_score(y_test, integer_predictions)

print(f"Accuracy with Integer Encoding: {integer_accuracy * 100:.2f}%")
```

```python
# Embedding Example

import numpy as np
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Dummy data
texts = [
    "This is a positive sentence.",
    "I love working with deep learning models.",
    "Negative reviews are not helpful.",
    "The movie was terrible; I hated it.",
]

labels = [1, 1, 0, 0]

# Tokenize the text for embedding
max_words = 10000
maxlen = 20
embedding_dim = 8

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
x_embedded = pad_sequences(sequences, maxlen=maxlen)

# Convert labels to NumPy array
y_labels = np.asarray(labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_embedded, y_labels, test_size=0.2, random_state=42, shuffle=True)

# Build the model with Embedding
embedding_model = Sequential()
embedding_model.add(Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=maxlen))
embedding_model.add(SimpleRNN(4))
embedding_model.add(Dense(1, activation='sigmoid'))
embedding_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with Embedding
embedding_model.fit(x_train, y_train, epochs=5, batch_size=2, validation_split=0.2, shuffle=True)

# Evaluate the model on the test set
embedding_predictions = (embedding_model.predict(x_test) > 0.5).astype('int32')
embedding_accuracy = accuracy_score(y_test, embedding_predictions)

print(f"Accuracy with Embedding: {embedding_accuracy * 100:.2f}%")
```

# Many-to-One RNN:

## In a Many-to-One RNN, the input sequence is of variable length, and the output is a fixed-size prediction. Here's an example using a simple LSTM layer:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

#In this example, the input shape is (None, 10), where None represents variable sequence length, and 10 is the number of features in each time step.
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10)))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# Many-to-Many RNN:

## In a Many-to-Many RNN, both the input and output sequences have variable lengths. Here's an example using the TimeDistributed layer to apply a Dense layer at each time step:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, TimeDistributed

#In this example, return_sequences=True in the LSTM layer ensures that it returns the full sequence of outputs for each input sequence.
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10), return_sequences=True))
model.add(TimeDistributed(Dense(1, activation='sigmoid')))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# One-to-Many RNN:

## In a One-to-Many RNN, the input is a fixed-size sequence, and the output is a variable-length sequence. Here's an example using RepeatVector:

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, RepeatVector

#In this example, the RepeatVector layer repeats the output of the previous layer to match the desired output sequence length.
output_sequence_length=10
model = Sequential()
model.add(LSTM(64, input_shape=(None, 10)))
model.add(RepeatVector(output_sequence_length))  # Set output_sequence_length accordingly
model.add(LSTM(64, return_sequences=True))
model.add(TimeDistributed(Dense(1, activation='sigmoid')))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

# Problems with RNN

Recurrent Neural Networks (RNNs) have been widely used for sequential data modeling, but they also have some limitations and challenges:

Vanishing and Exploding Gradients: RNNs suffer from vanishing and exploding gradient problems during training. This occurs because gradients are multiplied across many time steps during backpropagation, causing them to either become very small (vanishing gradients) or very large (exploding gradients). This can make it difficult for the model to learn long-term dependencies.

Difficulty in Capturing Long-Term Dependencies: RNNs have difficulty in capturing long-term dependencies in sequences. This is because information from earlier time steps may gradually diminish as it propagates through the network, making it hard for the model to remember information from distant past time steps.

Limited Contextual Information: RNNs process sequences one element at a time, which limits their ability to capture contextual information from the entire sequence. This may result in suboptimal performance, especially when dealing with long sequences.

Computationally Intensive: Training RNNs can be computationally intensive, especially for long sequences and deep architectures. This can lead to longer training times and higher resource requirements.

Fixed-Length Inputs and Outputs: Traditional RNNs require fixed-length inputs and produce fixed-length outputs, which can be restrictive for tasks involving variable-length sequences.

Difficulty in Parallelization: Due to their sequential nature, RNNs are difficult to parallelize efficiently across time steps, limiting their scalability on hardware accelerators like GPUs.

Difficulty with Training Stability: Training RNNs can be challenging to stabilize, especially for deep architectures and long sequences. Various techniques such as gradient clipping, batch normalization, and careful weight initialization are often required to address this issue.

Despite these challenges, RNNs have been foundational in sequential data modeling and have paved the way for more advanced architectures like Long Short-Term Memory (LSTM) networks and Gated Recurrent Unit (GRU) networks, which aim to address some of these limitations. Additionally, attention mechanisms and transformer architectures have emerged as powerful alternatives for sequential data processing in recent years.
